```python
from Bio.Seq import Seq
from Bio.SeqUtils import gc_fraction

def analyze_dna(sequence):
    sequence = sequence.upper().replace(" ", "").replace("\n", "")

    print("\n DNA Sequence Analysis\n" + "-"*40)

    # Length
    length = len(sequence)
    print(f"📏 Sequence Length: {length} bp")

    # GC Content
    gc = gc_fraction(sequence) * 100
    print(f"GC Content: {gc:.2f}%")

    # Start Codon Check
    if sequence.startswith("ATG"):
        print(" Start Codon (ATG): Present")
    else:
        print(" Start Codon (ATG): Not found at beginning")

    # Stop Codon Check
    stop_codons = ["TAA", "TAG", "TGA"]
    if sequence[-3:] in stop_codons:
        print(f" Stop Codon ({sequence[-3:]}): Present at end")
    else:
        print(" Stop Codon: Not found at end")

    # Translation
    dna_seq = Seq(sequence)
    protein = dna_seq.translate(to_stop=True)

    print("\n Protein Sequence:")
    print(protein)

    print("\n Analysis Complete\n")


# 🧾 Input from user
dna_input = input("Enter DNA sequence:\n")

analyze_dna(dna_input)
```
